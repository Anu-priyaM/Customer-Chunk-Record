## Step 6 : Data Visualization

```python
# Assigning data to df variable
df = pd.DataFrame(data)

# Creating a pie chart for Card Type
df['Card Type'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Card Type',weight='bold',fontsize=14)
plt.legend(loc='upper right', fontsize=12, shadow=True,bbox_to_anchor=(2,1))  # Set legend properties
plt.show()
```

```python
# Creating pie chart for Geography
colors = ['gold', 'yellowgreen', 'lightcoral']
df['Geography'].value_counts().plot(kind='pie',autopct='%1.1f%%',colors=colors,startangle=90)
plt.title('Geography',weight='bold',fontsize=14)
plt.legend(loc='upper right', fontsize=12, shadow=True,bbox_to_anchor=(2,1))
plt.xticks(rotation=140)
plt.show()
```

```python
# Creating bar chart for top 10 Age 
plt.figure(figsize=(6,3))
df['Age'].value_counts().head(10).plot(kind='bar',color='#FF0099',fontsize=10,edgecolor='black',width=0.8)
plt.xlabel('Age',weight='bold',fontsize=14)
plt.ylabel('Count',weight='bold',fontsize=14)
plt.title('Top 10 Age',weight='bold',fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

```python
# Creating  bar chart of top 10 EstimatedSalary
plt.figure(figsize=(6,4))
df['EstimatedSalary'].value_counts().head(10).plot(kind='bar',color='#110099',edgecolor='black',width=0.8)
plt.xlabel('EstimatedSalary')
plt.ylabel('Count')
plt.title('Top 10 EstimatedSalary')
plt.xticks(rotation=90)
plt.show()
```

```python
# Creating line chart to know Point Earned
plt.figure(figsize=(8,4))
df.groupby('NumOfProducts')['Point Earned'].mean().plot(marker='o')
plt.xlabel('NumOfProducts',weight='bold',fontsize=14)
plt.ylabel('Average Point Earned',weight='bold',fontsize=14)
plt.title('Point Earned',weight='bold',fontsize=14)
plt.show()
```
